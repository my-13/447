# 1. Sympy practice

```python
# define variables 
var("b s a k_p k_i")
# alternatively, you can also do b = Symbol('b', real=True) etc. to be more
# specific and careful about what you write


# define P and C
P = b/(s+a)
C = k_p + 1/s * k_i

# print
print("P,C equals:")
P,C
```

```python
# define Gyv = P/(1+PC)
Gyv = P/(1+P*C)
print("Gyv equals:")
Gyv
```

```python
# simplify Gyv using .simplify()
Gyv = Gyv.simplify()
Gyv
```

```python
# put Gyv in standard canonical form, p/q 
Gyv = Gyv.cancel()
Gyv
```
