You are hired by a small venture-backed startup company as the Senior (i.e. only) Control Systems engineer tasked with designing the controller for a self-balancing scooter.  The startup has been slow to bring a product to market, and investors are getting anxious -- it is critically important that the team brings a product to market before summer.

Your project manager has made the following design decisions / specifications:

- the rider will specify a reference velocity r using a smartphone app;

- the scooter will measure its velocity $\dot{p}$;

- your controller will determine the torque command $\tau$ for motors attached to the wheels.

Before your boss will let you touch (i.e. conduct experiments with) the physical hardware, you must synthesize and analyze your controller using a simulation model. You did some Googling, and found this dynamical system that seems to approximate the self-balancing scooter. 

$$\left[\begin{matrix}M+m & -m\ell\cos\theta \\ -m\ell\cos\theta & J+m\ell^2\end{matrix}\right]\left[\begin{matrix}\ddot{p} \\ \ddot{\theta}\end{matrix}\right] + \left[\begin{matrix}c\dot{p}+m\ell\dot{\theta}^2\sin\theta \\ \gamma\dot{\theta}-mg\ell\sin\theta\end{matrix}\right] = \left[\begin{matrix}\tau R \\ 0\end{matrix}\right]$$

where: $p$ is the position of the scooter; $\theta$ is the angle of the rider; $M$ is the mass of the scooter; $m$, $J$, and $\ell$ are the rider's mass, moment-of-inertia, and leg length (~half of rider height); $c$ and $\gamma$ are coefficients of friction, $R$ is the radius of the wheels, and $g$ is the acceleration due to gravity.  Since you don't expect riders will lean more than $\pm 15$ degrees from vertical, you are justified in making the small-angle approximations $\cos\theta\approx 1, \sin\theta\approx\theta$.

The mechanical design team reports the following parameter estimates:  $M=40$ kg, $R=0.25$ m, $g=9.81$ m/sec^2, $c=0.7$, $\gamma=0.2$. You did some Googling to figure out the rider's parameters ($m,J,\ell$) -- for your target demographic of American "tweens", that is, kids that are 8--12 years old, you estimated that the averages are $m=40 kg$ , $J=0.2 kg/m^2$, $\ell=0.9 m$.

Upon linearizing the system about the equilibrium point $\dot{p}=r,\theta=0,\dot{\theta}=0, u= \tau / r$, you found that for states $x = [\dot{p},\theta,\dot{\theta}],$ the system can be approximated by $\dot{\delta x}=A\delta x + B\delta u, \delta y = C\delta x+ D\delta u$, where 

$A = \left[\begin{matrix} -cJ_t/\mu & m^2\ell^2 g/\mu & -\gamma \ell m/\mu \\ 0 & 0 & 1 \\ -c\ell m/\mu & M_t m g\ell/\mu & -\gamma M_t/\mu\end{matrix}\right]$

$B = \left[\begin{matrix} RJ_t/\mu \\ 0 \\ R\ell m/\mu\end{matrix}\right]$

$C = \left[\begin{matrix} 1 & 0 & 0\end{matrix}\right]$

$D = \left[\begin{matrix} 0 \end{matrix}\right]$

We define $M_t = M+m, J_t = J+m\ell^2, \mu = M_tJ_t-m^2\ell^2$
